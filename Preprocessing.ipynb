```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler,Binarizer,Normalizer,OneHotEncoder,LabelEncoder
```

```python
np.random.seed(10)
```

```python
a=np.random.normal(0,1,500)
b=np.random.normal(-2,2,500)
c=np.random.normal(5,3,500)
```

```python
df = pd.DataFrame({
    'A' : a,
    'B' : b,
    'C' : c
})
df
```

```python
df.plot.kde()
```

```python
scalar=StandardScaler()
scaled=scalar.fit_transform(df)
df=pd.DataFrame(scaled,columns=['a','b','c'])
```

```python
df.plot.kde()
```

```python
mmm=MinMaxScaler()
```

```python
mm=mmm.fit_transform(df)
df=pd.DataFrame(mm,columns=['c','u','t'])
```

```python
df.plot.kde()
```
